```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time_cuda = np.array([1.02, 1.43, 1.81, 2.21, 2.61, 3.00, 3.38])
time_cuda_s = np.array([1.88, 2.61, 3.35, 4.09, 4.83, 5.57, 6.3])
time_cpu = np.array([26.25, 36.68, 47.49, 57.65, 68.90, 79.61, 89.64])

time_cuda_32_16 = np.array([0.88, 1.21, 1.54, 1.87, 2.21, 2.54, 2.87])
time_cuda_16_16 = np.array([0.83, 1.12, 1.43, 1.73, 2.01, 2.31, 2.68])
```

```python
time_cuda_m = np.array([1.98, 2.52, 3.09, 3.65, 4.2, 4.78, 5.56])
time_cuda_m_s = np.array([3.03, 4.06, 5.07, 6.09, 7.29, 8.56, 9.88])
time_cpu_m = np.array([34.84, 48.77, 62.73, 76.71, 90.66, 104.7, 118.66])

time_cuda_m_32_16 = np.array([1.79, 2.26, 2.76, 3.24, 3.73, 4.23, 4.94])
time_cuda_m_16_16 = np.array([1.65, 2.07, 2.49, 2.94, 3.38, 3.81, 4.47])
```

```python
s = time_cpu / time_cuda
s_m = time_cpu_m / time_cuda_m
s_s = time_cpu / time_cuda_s
s_m_s = time_cpu_m / time_cuda_m_s
```

```python
plt.figure(figsize=(13, 8))
plt.rcParams['font.size'] = 15

x = np.arange(500, 1701, 200)

plt.plot(x, time_cpu / time_cuda_16_16, '.-', label='2 GPUs (16x16)')
plt.plot(x, s, '.-', label='2 GPUs')
plt.plot(x, s_s, '.-', label='1 GPU')
plt.plot(x, np.ones_like(s), '.-', label='no speedup')

plt.xlabel('Grid size')
plt.ylabel('Speedup')
plt.xticks(x)
plt.legend(loc='best')

plt.show()
```

```python
plt.figure(figsize=(13, 8))
plt.rcParams['font.size'] = 15

x = np.arange(500, 1701, 200)

plt.plot(x, time_cpu_m / time_cuda_m_16_16, '.-', label='2 GPUs (16x16)')
plt.plot(x, s_m, '.-', label='2 GPUs')
plt.plot(x, s_m_s, '.-', label='1 GPU')
plt.plot(x, np.ones_like(s_m), '.-', label='no speedup')

plt.xlabel('Grid size')
plt.ylabel('Speedup')
plt.xticks(x)
plt.legend(loc='best')

plt.show()
```
