```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = pd.read_csv('result_8_2000_2000.p', sep=' ', header=None)
```

```python
u = data.values[:, :-1]
```

```python
def phi(x, y):
    return 2 / (1 + x ** 2 + y ** 2)
```

```python
numOfPts = 2000
```

```python
def get_pt(i):
    t = 1.0 * i / (numOfPts - 1)
    return 2 * ((1 + t) ** 1.5 - 1) / (2 ** 1.5 - 1)
```

```python
Xs = [get_pt(i) for i in range(numOfPts)]
u_real = [[phi(x, y) for x in Xs] for y in Xs]

X, Y = np.meshgrid(Xs, Xs)
diff = np.abs(u_real - u)
```

```python
plt.rcParams.update({'font.size': 13})
fig = plt.figure(figsize=(15, 12))
ax = Axes3D(fig)

# surf = ax.plot_surface(X, Y, u, color='palevioletred', shade=False)
surf = ax.plot_surface(X, Y, u_real, color='mediumseagreen', shade=False)
# surf = ax.plot_surface(X, Y, diff, color='coral')
ax.set_zlim((0, 2))

ax.view_init(elev=50, azim=-50)

ax.set_xlabel('x', labelpad=10)
ax.set_ylabel('y', labelpad=10)
ax.set_zlabel('z', labelpad=15)

plt.savefig('report/pics/real.pdf', bbox_inches='tight')
plt.show()
```
