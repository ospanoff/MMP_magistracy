```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 4, 8, 16]
s1_cuda = [1.66, 2.84, 5.38, 9.27, 16.81]
s2_cuda = [1.98, 3.58, 6.47, 12.21, 21.34]
s1_mpi = [1, 1.95, 3.94, 7.6, 15.67]
s2_mpi = [1, 1.95, 3.89, 7.1, 14.87]

e1_cuda = np.array(s1_cuda) / x / s1_cuda[0]
e2_cuda = np.array(s2_cuda) / x / s2_cuda[0]
e1_mpi = np.array(s1_mpi) / x
e2_mpi = np.array(s2_mpi) / x
```

```python
plt.rcParams['font.size'] = 20
plt.figure(figsize=(15,8))

plt.plot(x, s1_cuda, 'o-', label='CUDA. Сетка 1000х1000')
plt.plot(x, s2_cuda, 'o-', label='CUDA. Сетка 2000х2000')
plt.plot(x, s1_mpi, 'o-', label='MPI. Сетка 1000х1000')
plt.plot(x, s2_mpi, 'o-', label='MPI. Сетка 2000х2000')

plt.xlabel('Количество видеокарт/процессоров')
plt.ylabel('Ускорение')
plt.legend(loc='best')

# plt.savefig('report/pics/cuda_s.pdf', bbox_inches='tight')
plt.show()
```

```python
plt.rcParams['font.size'] = 20
plt.figure(figsize=(15,8))

plt.plot(x, e1_cuda, 'o-', label='CUDA. Сетка 1000х1000')
plt.plot(x, e2_cuda, 'o-', label='CUDA. Сетка 2000х2000')
plt.plot(x, e1_mpi, 'o-', label='MPI. Сетка 1000х1000')
plt.plot(x, e2_mpi, 'o-', label='MPI. Сетка 2000х2000')

plt.xlabel('Количество видеокарт/процессоров')
plt.ylabel('Эффективность')
plt.legend(loc='best')

# plt.savefig('report/pics/cuda_e.pdf', bbox_inches='tight')
plt.show()
```
